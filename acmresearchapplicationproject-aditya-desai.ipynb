```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("../input/carsforsale/cars_raw.csv")

data.head()
```

```python
#Function used to remove "$" and "," from price. This will be used later to determine total cost of every car from each year
def format_price(price):
    formated_price = price.replace('$', '')
    if 'Not Priced' in formated_price:
        return float(0)
    elif ',' in formated_price:
        formated_price = float(formated_price.replace(',', ''))
    return float(formated_price)
data['Price'] = data['Price'].apply(lambda x: format_price(x))
data.head()
        
```

```python
by_year = data.Year.value_counts().sort_index(ascending=True)
by_year
ranking = by_year[:23]
ranking
```

```python
ranking.plot(kind = "bar")
```
